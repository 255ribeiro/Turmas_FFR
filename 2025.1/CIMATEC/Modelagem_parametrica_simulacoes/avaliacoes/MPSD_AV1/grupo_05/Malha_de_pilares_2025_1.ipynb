```python
#Importando as bibliotecas.
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definindo a variação de x: nesse caso, x está entre -10 e 11, pulando de 0,2 em 0,25.
x = np.arange(-10, 11, .25)
x
```

```python
#Criando gráfico de linha baseado no seno de X.
plt.plot(np.sin(x))
```

```python
#Criando gráfico de linha e de pontos baseado no seno de X.
fig  , ax =  plt . subplots ()
ax.plot (x, np.sin(x))
ax.scatter (x, np.sin(x), c='red')
```

```python
#Definindo a distância entre os pilares e a quantidade deles em cada eixo.
dx = 10
ny = 3
dy = 10
```

```python
#Definindo as coordenadas de X e Y.
coord_x = []
coord_y = []

for i in range(nx):
    for j in range(ny):
        coord_x.append(i*dx)
        coord_y.append(j*dy)

print("x: ", coord_x)
print("y: ", coord_y)
```

```python
#Montando a malha com os pilares posisionados com a distância regular e adicionando um título à malha.
fig, ax = plt.subplots()
ax.scatter(coord_x, coord_y)
ax.set_aspect('equal')
plt.title ("malha regular")
```
