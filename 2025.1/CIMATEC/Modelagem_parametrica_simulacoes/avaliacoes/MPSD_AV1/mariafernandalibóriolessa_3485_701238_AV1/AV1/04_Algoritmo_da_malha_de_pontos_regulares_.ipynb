```python
import numpy as np #Importa a biblioteca NumPy e a apelida como np (usada para operações numéricas)
import matplotlib.pyplot as plt #Importa a biblioteca Matplotlib para geração de gráficos
```

```python
#Cria um vetor x que vai de -10 até 10 (exclusivo), com passo de 0.25
x = np.arange(-10, 10, .25)
x #Mostra o vetor x (em um ambiente interativo como Jupyter)
```

```python
#Plota apenas a curva do seno de x (sem especificar eixo x)
plt.plot(np.sin(x))
#Eixo x é o índice do vetor; eixo y é o seno dos valores de x
```

```python
#Cria uma nova figura e eixos (mais organizado para múltiplos gráficos)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x)) #Plota a curva do seno de x
ax.scatter(x, np.sin(x), c='red') #Adiciona pontos vermelhos sobre a curva do seno
```

```python
nx = 8 #Número de colunas (pontos no eixo x)
dx = 12.5 #Distância entre os pontos no eixo x
ny = 12 #Número de linhas (pontos no eixo y)
dy = 7.5 #Distância entre os pontos no eixo y
```

```python
#Inicializa listas para armazenar as coordenadas x e y dos pontos da malha
coord_x = []
coord_y = []

#Loop duplo para gerar as coordenadas de uma malha regular
for i in range(nx): #Para Cada coluna
  for j in range(ny): #Para cada linha
    coord_x.append(i*dx) #Calcula a posição x e adiciona à lista
    coord_y.append(j*dy) #Calcula a posição y e adiciona à lista

  print("x: ", coord_x) #Mostra as coordenadas x até o momento
  print("y: ", coord_y) # Mostra as coordenadas y até o momento
```

```python
#Cria uma nova figura com eixos
fig, ax = plt.subplots()
ax.scatter(coord_x, coord_y) #Plota os pontos da malha como um gráfico de dispersão
ax.set_aspect('equal') #Garante que a escala dos eixos x e y seja a mesma
plt.title("malha regular") #Adiciona título ao gráfico
```
