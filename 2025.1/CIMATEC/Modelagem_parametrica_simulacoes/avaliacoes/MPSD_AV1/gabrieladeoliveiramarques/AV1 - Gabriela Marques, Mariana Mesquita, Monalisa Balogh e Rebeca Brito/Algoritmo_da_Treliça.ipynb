```python
# AV1 - Modelagem Paramétrica

# GRUPO:
# - Gabriela Marques
# - Mariana Mesquista
# - Monalisa Balogh
# - Rebeca Brito

import matplotlib.pyplot as plt

# Definindo os parâmetros da treliça
num_modulos = int(input("Digite o número de módulos da treliça: "))
largura_modulo = float(input("Digite a largura de cada módulo (em metros): "))
altura_trel = float(input("Digite a altura da treliça (em metros): "))

# Listas para armazenar os nós (pontos) superiores e inferiores
nos_superiores = []
nos_inferiores = []

# Criando os nós da treliça
for i in range(num_modulos + 1):
  x = i * largura_modulo
  nos_superiores.append((x, altura_trel))     #Banzo superior
  nos_inferiores.append((x, 0))               #Banzo inferior

 # Criando o gráfico
plt.figure(figsize=(10, 4))

 # Desenhando os banzos (ligação entre nós)
for i in range(num_modulos):

     # Banzo inferior
     x_vals = [nos_inferiores[i][0], nos_inferiores[i + 1][0]]
     y_vals = [nos_inferiores[i][1], nos_inferiores[i + 1][1]]
     plt.plot(x_vals, y_vals, 'k-')

     # Banzo superior
     x_vals = [nos_superiores[i][0], nos_superiores[i + 1][0]]
     y_vals = [nos_superiores[i][1], nos_superiores[i + 1][1]]
     plt.plot(x_vals, y_vals, 'k-')

     # Diagonal (cruzando entre os banzos)
     if i % 2 == 0:
      plt.plot(
          [nos_inferiores[i][0], nos_superiores[i + 1][0]],
          [nos_inferiores[i][1], nos_superiores[i + 1][1]],
      'b--'
      )

     else:
      plt.plot(
        [nos_superiores[i][0], nos_inferiores[i + 1][0]],
        [nos_superiores[i][1], nos_inferiores[i + 1][1]],
        'b--'
    )

# Adicionando nós como pontos
for (x, y) in nos_superiores + nos_inferiores:
  plt.plot(x, y, 'ro') # Pontos vermelhos

# Exibindo a treliça
  plt.title("Treliça com Banzos e Diagonais")
  plt.axis("equal")
  plt.grid(True)
  plt.xlabel("X (m)")
  plt.ylabel("Y (m)")
  plt.show()
```
