```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# entradas
comp = 10.0 #Float
h = 2.0 #Float
n_nos = 10 # int
```

```python
coord_x = np.linspace(0, comp, n_nos)
coord_y1 = [0]* n_nos
coord_y2 = [h]* n_nos
```

```python
diag_y = []
for i in range (n_nos):
    if i %2 == 0:
        diag_y.append(0)
    else:
        diag_y.append(h)
print(diag_y)
```

```python
fig, ax = plt.subplots()

ax.set_title("Treliça") #título da treliça
ax.scatter(coord_x, coord_y1) #desenho dos pontos inferiores
ax.plot (coord_x, coord_y1) #desenho das linhas inferiores

ax.scatter(coord_x, coord_y2) #desenho dos pontos superiores
ax.plot (coord_x, coord_y2) #desenho das linhas superiores

ax.plot (coord_x, diag_y)

ax.axis('off') #apagar desenho dos eixos
ax.set_aspect('equal') #deixar oo tamanho proporcional
```
