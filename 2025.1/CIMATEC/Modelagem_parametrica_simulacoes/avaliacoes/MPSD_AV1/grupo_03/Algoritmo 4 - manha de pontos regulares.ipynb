```python
import numpy as np
```

```python
nx = 4
dx = 10
ny = 3
dy = 10
```

```python
import matplotlib.pyplot as plt

def gerar_malha_pontos(dx, dy, nx, ny):
    """
    Gera uma imagem de uma malha de pontos regulares.

    Parâmetros:
    - dx: distância entre pontos no eixo x
    - dy: distância entre pontos no eixo y
    - nx: número de pontos no eixo x
    - ny: número de pontos no eixo y
    """
    # Coordenadas dos pontos
    x_coords = [i * dx for i in range(nx)]
    y_coords = [j * dy for j in range(ny)]

    # Criar figura
    plt.figure(figsize=(6, 6))

    # Gerar os pontos
    for x in x_coords:
        for y in y_coords:
            plt.plot(x, y, 'ko')  # 'ko' = marcador circular preto

    # Ajustar limites e aspecto do gráfico
    plt.xlim(-dx, dx * nx)
    plt.ylim(-dy, dy * ny)
    plt.gca().set_aspect('equal')  # manter proporção entre eixos
    plt.title('Malha de Pontos Regulares')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, linestyle='--', alpha=0.3)

    # Exibir a malha
    plt.show()

# Exemplo de uso
gerar_malha_pontos(dx=1, dy=1, nx=10, ny=10)
```
